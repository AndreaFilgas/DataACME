```python
rt_info = pd.read_csv('C:/Users/afilg/OneDrive/Área de Trabalho/DATA SCIENCE/Aulas - Originais/Projeto_Integrador/Dados/rt_movie_info.csv', encoding = 'latin1', engine = 'python', delimiter = ',')
rt_info.head().T
```

```python
rt_info_critica = rt_info.join(rt_review, on = 'movie_link', how = 'inner')
```

```python
rt_info_critica = rt_info_critica.dropna(subset=['in_theaters_date'])
```

```python
rt_info_critica['year'] = rt_info_critica['in_theaters_date'].str[-4:].astype(np.int)
```

```python
rt_info_critica['movieyear'] = rt_info_critica.year.astype(str) + rt_info_critica.movie_name
```

```python
rt_info_critica.drop_duplicates(subset ="movieyear", keep = False, inplace = True)
```
